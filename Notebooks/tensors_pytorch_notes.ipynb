# Tensors Basic Notes

### Imports

```python
import torch
import numpy as np
```

### Tensors

Tensor - A high dimension array, the parent of smaller ones like the matrix, vector, or scalar

```python
#Initialize a tensor

data = [[1,2],[3,4]]
x_data = torch.tensor(data) #Turns data into a tensor object
print(x_data)
```

**From Numpy:** Tensors can also be created from a numpy array.

```python
np_array = np.array(data)
x_np = torch.from_numpy(np_array)
print(x_np)
```

Tensors can also be created from other tensors and data, like creating a tensor full of ones of the same size, or a random tensor.

```python
x_ones = torch.ones_like(x_data) # retains the properties of x_data
print(f"Ones Tensor: \n {x_ones} \n")

x_rand = torch.rand_like(x_data, dtype=torch.float) # overrides the datatype of x_data
print(f"Random Tensor: \n {x_rand} \n")
```

**Shape:**  This is a tuple that determines dimensionality

```python
#Each input consecutively adds a new dimension, so last value will be the length of a single array and it builds from there.
shape = (2,2,4,)
ones_tensor = torch.ones(shape)

print(f"Ones Tensor: \n {ones_tensor} \n")
```

### Attributes

**Attributes inlude:** Shape, data type, and the device it is stored on

```python
shape = (3,4)
tensor = torch.rand(shape)

print(f"Shape of tensor: {tensor.shape}")
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")
```

## Operations

There are lots and lots of operations :(
All of these can run on gpu or cpu so you have to ensure its on the one you want

```python
#Move to gpu if availible
if torch.accelerator.is_available():
    tensor = tensor.to("cuda")
    # can alo do: tensor.to(torch.accelerator.current_accelerator())
    print(f"Device tensor is stored on: {tensor.device}")
```

```python
#Explore a few basic operations
tensor = torch.ones(4,4)
print(tensor)
print()

# Indexes
print(f"Row by first index: {tensor[0]}")
print(f"Column by first index: {tensor[:,0]}")
print(f"Row by last index: {tensor[-1]}")
print(f"Column by last index: {tensor[...,-1]}")
print()

#Altering rows and columns
tensor[:, -1] =0
tensor[2] = 0
print(tensor)
print()

#Concat - combine along a given dimension (torch.stack is another option)
combined = torch.cat([tensor, tensor, tensor], dim=1)
print(combined)
```

```python
# Matrix multiplication. - tensor.T returns a transpose of a tensor (flip rows and columns)
print(tensor)
print()
print(tensor.T)
print()
y1 = tensor @ tensor.T
y2 = tensor.matmul(tensor.T)
print(y1)
print()
print(y2)
```

```python
# Element multiplication
tensor2 = torch.ones(4,4)

mult = tensor * tensor2 # Can also use mult function
print(mult)
```

Sum and other tensor functions return a condensed tensor object. Using .item() gets the item and returns the desired value of the operation.

```python
#Items and sum
agg = tensor.sum()
agg_item = agg.item()
print(agg)
print(agg_item)
```

#### Important Note

  - any operation ending in _ acts as in place like in matplotlib/pandas, applying the change to the tensor.
